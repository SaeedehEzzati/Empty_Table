```python
import cv2
import threading
import numpy as np
from queue import Queue
from datetime import datetime
import time
from threading import  Thread

queue = Queue()

RECONNECTION_PERIOD = 5

class ThreadedCamera(object):
    def __init__(self,source='cam:Sayda@0.0.0.0:88/videoMain'):
       
        self.source = source
        self.connected_camera_flag = False
        self.capture,self.capture_flag = self.Connect_Camera(True)
        
#         print(self.capture_flag)
        
        if self.capture_flag:
            self.thread = Thread(target=self._Capture_Read)
            self.thread.daemon = True
            self.thread.start()
        
        else: return None

    def _Capture_Read(self):
        try:
#             print(1)
            counter = 0
            while True:
#                 print(2)
                grabbed = False
                while not grabbed:
#                     print(3)
                    if self.capture is None:
#                         print(4)
                        pass
                    else: 
#                         print(5)
                        grabbed = self.capture.grab()
#                         print(55)
                        
                    if not grabbed:
#                         print(6)
                        pass
                    else: 
#                         print(7)
                        counter += 1

                if counter > 10:
#                     print(8)
                    self.capture,self.capture_flag = self.Connect_Camera(False)
                    grabbed = self.capture_flag
#                     time = datetime.now()
#                     time.sleep(self.FPS)
                else: pass
        except KeyboardInterrupt: self.End_Frame()
    
    
    def Connect_Camera(self,enum):
        counter = 0
        while True:
            capture = cv2.VideoCapture(f'rtsp://{self.source}')
#             capture = cv2.VideoCapture(self.source)
            if capture is None: 
                print('capture is None')
#                 time = datetime.now()
                time.sleep(RECONNECTION_PERIOD)
                counter += 1
                if counter > 10:
                    if self.connected_camera_flag:
                        print('self.connected_camera_flag')
                        enum = True
                        self.connected_camera_flag = False
                        
                    else:
                        print(None,False)
                        return None,False

                else: continue
            else:
                if enum:
#                     time = datetime.now()
                    self.connected_camera_flag = True
                else: 
                    pass

                return capture,True

    def Capture(self):
        if self.capture_flag:
            flag,frame = self.capture.retrieve()
#             print(flag)
            if flag : 
#                 print("put in queue")
                queue.put(frame)
            else:
                pass
#             cv2.waitKey(1)
        else: print("<< No Frame >>")
    
    def End_Frame(self):
        self.capture.release()

def run_cameras_forever(src):
   
    streamer = ThreadedCamera(src)
    count = 0
    while True:
#         print(queue.qsize())
        if streamer is None: 
#             print("streamer is none")
            continue
        else: pass
        
        streamer.Capture()
        if queue.empty(): 
#             print("q is empty")
            continue
        else: 
#             print("get from queue")
            print(queue.qsize())
            img = queue.get()

        if img is None: 
#             print('img is none')
            continue
        else: pass
        count = count + 1
#         print(count)
        img = cv2.resize(np.array(img), (900, 540))
        img = cv2.cvtColor(np.array(img),cv2.COLOR_BGR2RGB)
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        cv2.imshow('img', img)
        
        if cv2.waitKey(1) & 0xFF == ord('q'): #  frame_counter == 18:#
            cv2.destroyAllWindows()
            streamer.End_frame()
            break


if __name__ == '__main__':
    f'rtsp://{self.source}'src = 'rtsp:Sayda@0.0.0.0'
#     src = './samples/0/sample_3_lable_-1_2.mp4'
    t = threading.Thread(target=run_cameras_forever(src))
    t.setDaemon(True)
    t.start()
```
